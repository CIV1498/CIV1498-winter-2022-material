```python
# Import 3rd party libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

# Configure Notebook
%matplotlib inline
plt.style.use('fivethirtyeight')
sns.set_context("notebook")
import warnings
warnings.filterwarnings('ignore')
```

# NBA Dataset

```python
nba = pd.read_csv('nba.csv')[['AST', '3PA', 'PTS']]
nba.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = nba[['AST']]
y = nba[['PTS']]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=0
)
```

```python
ax = sns.scatterplot(X_train['AST'], y_train['PTS'], label='Train')
sns.scatterplot(X_test['AST'], y_test['PTS'], label='Test', ax=ax)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.set_xlabel('Average Assists Per Game, AST', fontsize=14)
ax.set_ylabel('Average Points Per Game, PTS', fontsize=14)
ax.legend(fontsize=12)
plt.show()
```
